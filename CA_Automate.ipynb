```python
lstProductType = ['Accessory Type','Bag Type','Braces Type','Bump Type','Clothing  Type','Coat Type','Dress type',
                  'Hat Type','Hosiery Type','Sandal Type','Shirt Type','Shoe Type','Skirt Type','Sock Type','Swimwear Type',
                  'Trouser Type','Underwear Type','Uniform Type']  
lstsize = ['Size Type','Unisex Sizes (S,M,L)']
```

```python
def add_multiselect(df):
    for col in df.columns:
        #25 upto y
        if col.strip() == 'Activity':
            df.rename(columns={col:'Activity -> multiselect'}, inplace=True)
        elif col.strip() == 'Amenities Included':
            df.rename(columns={col:'Amenities Included -> multiselect'}, inplace=True)
        elif col.strip() == 'Comfort':
            df.rename(columns={col:'Comfort -> multiselect'}, inplace=True)
        elif col.strip() == 'Compatible Model':
            df.rename(columns={col:'Compatible Model -> multiselect'}, inplace=True)
        elif col.strip() == 'Compatible With':
            df.rename(columns={col:'Compatible With -> multiselect'}, inplace=True)
        elif col.strip() == 'Compliance':
            df.rename(columns={col:'Compliance -> multiselect'}, inplace=True)
        elif col.strip() == 'Complications':
            df.rename(columns={col:'Complications -> multiselect'}, inplace=True)
        elif col.strip() == 'Delivery Times':
            df.rename(columns={col:'Delivery Times -> multiselect'}, inplace=True)
        elif col.strip() == 'Design Features':
            df.rename(columns={col:'Design Features -> multiselect'}, inplace=True)
        elif col.strip() == 'Entertainment':
            df.rename(columns={col:'Entertainment -> multiselect'}, inplace=True)
        elif col.strip() == 'Exterior Design':
            df.rename(columns={col:'Exterior Design -> multiselect'}, inplace=True)
        elif col.strip() == 'Features':
            df.rename(columns={col:'Features -> multiselect'}, inplace=True)
        elif col.strip() == 'Features / Comfort':
            df.rename(columns={col:'Features / Comfort -> multiselect'}, inplace=True)
        elif col.strip() == 'Food Specifications':
            df.rename(columns={col:'Food Specifications -> multiselect'}, inplace=True)
        elif col.strip() == 'Frequency Bends':
            df.rename(columns={col:'Frequency Bends -> multiselect'}, inplace=True)
        elif col.strip() == 'Interior Design':
            df.rename(columns={col:'Interior Design -> multiselect'}, inplace=True)
        elif col.strip() == 'Opening Days':
            df.rename(columns={col:'Opening Days -> multiselect'}, inplace=True)
        elif col.strip() == 'Opening Hours':
            df.rename(columns={col:'Opening Hours -> multiselect'}, inplace=True)
        elif col.strip() == 'Safety':
            df.rename(columns={col:'Safety -> multiselect'}, inplace=True)
        elif col.strip() == 'Skin Type':
            df.rename(columns={col:'Skin Type -> multiselect'}, inplace=True)
        elif col.strip() == 'Smart Capability':
            df.rename(columns={col:'Smart Capability -> multiselect'}, inplace=True)
        elif col.strip() == 'Special Features':
            df.rename(columns={col:'Special Features -> multiselect'}, inplace=True)
        elif col.strip() == 'Supported Modes':
            df.rename(columns={col:'Supported Modes -> multiselect'}, inplace=True)
        elif col.strip() == "What's Included":
            df.rename(columns={col:"What's Included -> multiselect"}, inplace=True)
        elif col.strip() == "Suitable For":
            df.rename(columns={col:"Suitable For -> multiselect"}, inplace=True)

    return df
```

```python
import os
import pandas as pd
import numpy  as np

######## Update the name for every category. This goes to header and attribute file names ######### 
catname = 'CA'

################ Get file and folder paths ################
currdir=os.getcwd()
infolder = os.path.join(currdir, 'input_files')
attfilename = "Attributes_" + catname + ".xlsx"
endattfilename = "EndAttributes_" + catname + ".xlsx"
headerfilename = "Headers_" + catname + ".xlsx"
outpath = currdir + "\\" + "Out_files"

########### Create output directory if not exists ###########
if not os.path.isdir(outpath):
    os.mkdir(outpath)

############ Load Header file and Attribute file ###########################
dfatt = pd.read_excel(currdir +  "\\" + attfilename,header=0)
dfendatt = pd.read_excel(currdir +  "\\" + endattfilename,header=0)
dfhead = pd.read_excel(currdir +  "\\" + headerfilename, header = 0)

```

```python
prodfilename = os.listdir(infolder)
i=0
#Check if all the entries in header file has corresponding entry
for index,row in dfhead.iterrows():
    header = row.tolist()[0]
    if  header not in prodfilename:
        print(header)
```

# All Files


```python
for prodfilename in os.listdir(infolder):
    #############################################
    #if "Pr_" + prodfilename in os.listdir(outpath):
    #    continue
    ###########################################
    print(prodfilename)
    ptcnt = 1
    
    ############# Read Product file ############################
    dfprod = pd.read_excel(infolder +  "\\" + prodfilename)
    #delete first col if file is 'Baby'
    if dfprod.columns[0] == 'Unnamed: 0':
        del dfprod[dfprod.columns[0]]
    ############ Create breadcrumbs dataframe ###################################
    dfprodbc = dfprod[dfprod.columns[0]]

    #Delete breadcrumbs header
    del dfprod[dfprod.columns[0]]

    #Remove spaces at head and tail of headers
    dfprod = dfprod.rename(columns=lambda x: x.strip())

    
    ###################### Get header list ######################
    #Standard headers starts from col 7
    stdheadstart = 7
    lstfromhead=[]

    for index,row in dfhead.iterrows():
        lsthead = row.tolist()
        if lsthead[0] == prodfilename:
            lstfromhead = lsthead[stdheadstart:]    #Get columns other than breadcrumbs
            lstfromhead = [x for x in lstfromhead if str(x) != 'nan'] #remove nan
            #We got the necessary data, just break the loop
            break
            
    if not lstfromhead:
        print('********* ABOVE FILE NOT PRESENT IN HEADER LIST************* ')
        continue
    
    lstbc  = lsthead[1:stdheadstart]
            
    for  colx in dfprod.columns :
        #Reform the headers containing  '/' with spaces before and after /
        if '/' in colx:
            temp = colx.split('/')
            col = temp[0].strip() + ' / ' + temp[1].strip()
            dfprod.rename(columns={colx:col}, inplace=True)
        else:
            col = colx.strip()

        #If 'What's Included' and 'Material' present in dfprod, update the values in dfatt
        if col == "What's Included":
            dfendatt["What's Included"] = dfprod[colx]
        elif col == "Material":    
            dfendatt["Material"] = dfprod[colx]

        if col == 'From':
            dfprod.rename(columns={col:'To'}, inplace=True)
        elif col == 'Age':
            dfprod.rename(columns={col:'Age Range'}, inplace=True)
        elif col == 'Bust Size':
            dfprod.rename(columns={col:'Chest Size'}, inplace=True)
        elif col == 'Closure':
            dfprod.rename(columns={col:'Closure Type'}, inplace=True)
        elif col == 'Specific Features' or col == 'Design Features':
            dfprod.rename(columns={col:'Features'}, inplace=True)
        elif col == 'Shoe Height':
            dfprod.rename(columns={col:'Heel Height'}, inplace=True)
        elif col == 'Ladies Shoe Size':
            dfprod.rename(columns={col:'Shoe Size'}, inplace=True)
        elif col == 'Fitted Sizes' or col == 'Clothing Sizes':
            dfprod.rename(columns={col:'Sizes'}, inplace=True)
        elif col == 'Theme':
            dfprod.rename(columns={col:'Subject'}, inplace=True)
        elif col == 'Shoe Width':
            dfprod.rename(columns={col:'Width'}, inplace=True)
        elif col in lstProductType:
            if 'Product Type' not in dfprod.columns:
                dfprod.rename(columns={col:'Product Type'}, inplace=True)
            else:
                if col in lstfromhead:
                    lstfromhead.remove(col)
                #dfprod.rename(columns={col:'Product Type(' + str(ptcnt) + '-' + col + ')'}, inplace=True)
                indp = lstfromhead.index('Product Type')
                #lstfromhead.insert(indp+1,'Product Type(' + str(ptcnt) + '-' + col + ')')
                lstfromhead.insert(indp+1,col)
        elif col in lstsize:
            if 'Sizes (S,M,L)' not in dfprod.columns:
                dfprod.rename(columns={col:'Sizes (S,M,L)'}, inplace=True)
            else:
                if col in lstfromhead:
                    lstfromhead.remove(col)
                #dfprod.rename(columns={col:'Product Type(' + str(ptcnt) + '-' + col + ')'}, inplace=True)
                if 'Sizes (S,M,L)' in lstfromhead:
                    indp = lstfromhead.index('Sizes (S,M,L)')
                    #lstfromhead.insert(indp+1,'Product Type(' + str(ptcnt) + '-' + col + ')')
                    lstfromhead.insert(indp+1,col)

       
    #Drop all the columns that are present in attributes file. 
    for c in dfprod.columns:
        if c in dfatt.columns or c in dfendatt.columns:
            del dfprod[c]
      
    
    ############ New attribute dataframe will have only necessary headers #################
    dfattnew=pd.DataFrame()
    #Keep only columns that are in headers file
    for i in lstfromhead:
        for col in dfatt.columns:
            if col.strip() == i.strip():
                dfattnew[col] = dfatt[col]

    ################### Concat attributes and product dataframes  ###########################
    dfunsorted = pd.concat([dfattnew,dfprod],axis=1)
    ############################# Re-order the columns ######################################
    dfunsorted = dfunsorted.reindex(columns  = lstfromhead)

    ############### If product type or doll type doesnt have any values fill it with last element from breadcrumb ##################
    lstnotnull = [x for x in lstbc if str(x) != 'nan']
    lstnotnull = [x for x in lstnotnull if str(x).strip() != '']
    
    if 'Product Type' in dfunsorted:
        if not dfunsorted['Product Type'].isnull().values.all():
            
            if str(dfunsorted['Product Type'] [0])=='nan' or str(dfunsorted['Product Type'] [0]).strip()=='' or str(dfunsorted['Product Type'][0]).lower()=='not specified':
                
                dfunsorted['Product Type'][0] = lstnotnull[-1]
        else:
            
            dfunsorted['Product Type'][0] = lstnotnull[-1]
            
    dfcombined = pd.concat([dfunsorted,dfendatt],axis=1)
    for col in dfcombined.columns:

        #If a column does not have any value print'Not Specified'
        if dfcombined[col].isnull().values.all() :
            dfcombined.iloc[0,dfcombined.columns.get_loc(col)] = "Not Specified"
        else:
            #For all the columns make sure 'Not Specified' is last entry
            lstns = dfcombined[col].values.tolist()
            lstns = [x for x in lstns if str(x) != 'nan']

            if not str(lstns[-1]).strip().lower() == "not specified":
                if len(lstns) == dfcombined.shape[0]:
                    dfcombined = dfcombined.append(pd.Series(), ignore_index=True)
                dfcombined.iloc[len(lstns),dfcombined.columns.get_loc(col)] = "Not Specified"


    #Add breadcrumbs header,dfprod and dfatt together 
    dffinalprod = pd.concat([dfprodbc,dfcombined],axis=1) 

    #Rename 'multiselect' columns
    dffinalprod = add_multiselect(dffinalprod)

    #Write to new excel
    dffinalprod.to_excel(outpath + "\\Pr_" + prodfilename,index=False)
print('done!')
```
