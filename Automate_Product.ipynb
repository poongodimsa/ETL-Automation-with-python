# Product file automation

```python
import os
import pandas as pd

currdir=os.getcwd()

catname = 'Clothes_Shoes_Accessories'
foldername = os.path.join(currdir, catname)
attfilename = "Attributes_" + catname + ".xlsx"
headerfilename = "Headers_" + catname + ".xlsx"
outpath = currdir + "\\" + "Out_files"

#Create output directory
if not os.path.isdir(outpath):
    os.mkdir(outpath)

#Get fixed attributes list
dfatt = pd.read_excel(currdir +  "\\" + attfilename,header=0)
```

```python

for prodfilename in os.listdir(foldername):
    #Read product file to df
    dfprod = pd.read_excel(foldername +  "\\" + prodfilename)
    #Do not proceed for 'Baby products'
    if dfprod.columns[0].find('Baby') == -1:

        #Delete breadcrumbs header
        del dfprod[dfprod.columns[0]]

        #Drop all the columns that are present in attributes file. And also the other cols such as 'period/Era' and 'Material'
        for  col in dfprod.columns :
            if (col.strip() in dfatt.columns or col.strip() == "Period / Era" 
                or col.strip() == "Colour" or 
                col.strip() == "Colours" or col.strip() == "Product Line" or col.strip() == "Material"):
                del dfprod[col]
            elif col.strip() == 'Designer / Brand':
                dfprod.rename(columns={col:'Brand'}, inplace=True)
            elif col.strip() == 'Model / No.':
                dfprod.rename(columns={col:'Model'}, inplace=True)
            elif col.strip() == 'Type':
                if 'Product Type' in dfprod.columns:
                    del dfprod[col]      #Delete the col if there is already a product type col
                else:
                    dfprod.rename(columns={col:'Product Type'}, inplace=True)

        #Concat attributes and product dataframes
        dfunsorted = pd.concat([dfatt,dfprod],axis=1)
        
        
        #Get the contents of HEADER file , reorder the columns as per requirement
        dfhead = pd.read_excel(currdir +  "\\" + headerfilename,header=0)

        #Standard headers starts from col 7
        stdheadstart = 7
        for index,row in dfhead.iterrows():
            lsthead =row.tolist()
            if lsthead[0] == prodfilename:

                #Get dataframe with 
                lstfromhead = lsthead[stdheadstart:-1]    #Get columns other than breadcrumbs
                lstfromhead = [x for x in lstfromhead if str(x) != 'nan'] #remove nan
                #We got the necessary data, just break the loop
                break

        #Reorder the columns
        dfunsorted = dfunsorted.reindex(columns  =lstfromhead)

        #If a column does not have any value print'Not Specified'
        for col in dfunsorted.columns:
            if dfunsorted[col].isnull().values.all():
                dfunsorted[col][0] = "Not Specified"

        #Create breadcrumbs dataframe
        lstbc  = lsthead[1:stdheadstart] 
        dfprodbc = pd.DataFrame(columns = lstbc)

        #Add breadcrumbs header and form the final dataframe
        dffinalprod = pd.concat([dfprodbc,dfunsorted],axis=1) 

        #Write to new excel
        dffinalprod.to_excel(outpath + "\\" + prodfilename,index=False) 

```
